### (Predicting the percentage of an student based on the number of study hours)

## Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

## Importing dataset and then cleaning it if required

```python
ds=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
ds.head()
```

```python
ds.isna().sum()
```

```python
ds.isnull().sum()
```

```python
ds.describe()
```

## let's check if the dataset contains any outliers or not

```python
ds.boxplot()
```

# Data Visualization

```python
ds.plot(x='Hours', y='Scores', style='o')
plt.xlabel("Hours of study")
plt.ylabel("Percentage of marks")
plt.title("Hours VS Percentage of marks")
plt.show()
```

```python
X,Y=ds.iloc[:,:-1].values,ds.iloc[:,1].values
```

### Spliting the data for Training and Testing purpose

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=32)
```

## Training of the algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regressor=LinearRegression()
linear_regressor.fit(X_train,Y_train)
```

```python
# Plotting the regression line
line = linear_regressor.coef_*X+linear_regressor.intercept_
plt.scatter(X, Y)
plt.plot(X, line, c='green')
plt.show()
```

```python
y_pred=linear_regressor.predict(X_test)
y_pred
```

```python
Y_test
```

### Predicted score if a student studies for 9.25 hrs/day

```python
hr=[[9.25]]
pred_Ans=linear_regressor.predict(hr)
pred_Ans
```

# Evaluation of the algorithm

```python
from sklearn import metrics
```

```python
print("Mean absolute error: ",metrics.mean_absolute_error(Y_test, y_pred))
```
