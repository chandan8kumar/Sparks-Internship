### (Predicting the class of Iris-Flower using kmeans method)

## Importing necessary libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn
```

## Importing the iris dataset

```python
from sklearn import datasets
iris_ds=datasets.load_iris()
```

```python
df=pd.DataFrame(iris_ds.data, columns=iris_ds.feature_names)
x = df.iloc[:, [0,1,2,3]].values
df.head()
```

## Using the elbow method to determine the number of clusters best for the iris dataset and then applying K-Means method

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,10),wcss,"-o")
plt.grid(True)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

## From above graph we can infer that k=3 is best optimum choice

```python
kmeans=KMeans(n_clusters=3)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

## Visualising the clusters on the basis of first two features

```python
plt.figure(figsize=(14,10))
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1],s=80,c='red', label='Setosa')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1],s=80,c='blue', label='Versicolour')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1],s=80,c='green', label='Virginica')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=80, c='yellow', label='Centroids')
plt.legend()
plt.show()
```

# Conclusion
## We conclude that there are three clusters for the Iris-dataset
